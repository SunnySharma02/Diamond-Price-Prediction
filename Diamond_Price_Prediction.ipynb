#Import Library Pandas

```python
import pandas as pd
```

#Read Data

```python
data=pd.read_csv('https://www.dropbox.com/s/5saczr8wmjketen/DiamondPrices.csv?dl=1')
```

```python
data
```

#Importing OrdinalEncoder

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
data.loc[:,['cut','color','clarity']]=oe.fit_transform(data[['cut','color','clarity']])
```

#Define y and X

```python
y=data['price']
y.shape
```

```python
X=data[['cut','color','clarity','sn','carat','depth','table','x','y','z']]
X.shape
```

#train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=2529)
```

#Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
```

#Model Fitting

```python
model.fit(X_train,y_train)
```

#Model Testing

```python
y_pred=model.predict(X_test)
y_pred
```

#Accuracy Check

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,mean_absolute_percentage_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_test)
```
